## Q1. 用至少两种办法生成一个3 × 4二维NumPy数组，数组的每个值都是10。

```python
import numpy as np
array = np.array([[10, 10, 10, 10], [10, 10, 10, 10], [10, 10, 10, 10]])
array
```

```python
array_1 = np.ones((3, 4)) * 10
array_1
```

```python
array_2 = np.zeros((3, 4)) + 10
array_2
```

```python
array_3 = np.full((3, 4), 10)
array_3
```

## Q2. 利用numpy.meshgrid() 和 matplotlib.pyplot.contour() 绘制二元函数 $ f(x_1, x_2) = x_2\exp(-x_1^2 - x_2^2)$ 的平面等高线。

```python
x1_array = np.linspace(-3,3,121)
x2_array = np.linspace(-3,3,121)

xx1,xx2 = np.meshgrid(x1_array, x2_array)

ff = xx2 * np.exp(-xx1**2 - xx2**2)
```

```python
# 导入matplotlib.pyplot模块
import matplotlib.pyplot as plt

# 创建一个大小为5x5的图形
fig, ax = plt.subplots(figsize = (5,5))

# 用numpy.meshgrid()函数生成网格点坐标矩阵
# xx1和xx2是两个矩阵，每个矩阵的行数和列数都是121
xx1,xx2 = np.meshgrid(x1_array, x2_array)

# 定义二元函数
ff = xx2 * np.exp(-xx1**2 - xx2**2)

# 绘制等高线图
CS = ax.contour(xx1, xx2, ff,
           levels = 20,
           cmap = 'RdYlBu_r')

# 添加颜色条
fig.colorbar(CS)

# 设置x轴和y轴标签
ax.set_xlabel('$\it{x_1}$')
ax.set_ylabel('$\it{x_2}$')

# 隐藏x轴和y轴刻度
ax.set_xticks([])
ax.set_yticks([])

# 设置x轴和y轴的范围
ax.set_xlim(xx1.min(), xx1.max())
ax.set_ylim(xx2.min(), xx2.max())


# 设置x轴和y轴的比例相等
ax.set_aspect('equal', adjustable='box')
```

## Q3. 在 [0, 1] 范围内生成1000个满足连续均匀随机数，并用matplotlib.pyplot.hist()绘制频率直方图

```python
X = np.random.exponential(scale=1, size=(50, 10))

fig, ax = plt.subplots(figsize = (5,5))
ax.hist(X, 100, density=False)
```
